```python
soup = BeautifulSoup(response.content, "lxml")
table_text = str(soup.find("textarea").text)
print(table_text)
```

```python
lines = table_text.split("\n")
lines = lines[2:-19]
lines
```

```python
import pandas as pd
```

```python
columns = lines[0].split("|")
for j in range(len(columns)):
    columns[j] = str(columns[j]).strip()
df = pd.DataFrame(columns = columns)
```

```python
for line in lines[3:]:
    cells = line.split("|")
    for j in range(len(cells)):
        cells[j] = str(cells[j]).strip()
    df_line = pd.DataFrame([cells], columns = columns)
    df = pd.concat([df, df_line], ignore_index = True)
```

```python
df_new = df[["Term", "Definition"]]
df_new.loc[:,"Definition"] = df_new["Definition"].str.replace('\\"', '"')
df_new.loc[:,"Definition"] = df_new["Definition"].str.replace("\\'", "'")
df_new.style.set_properties(**{'white-space': 'pre-wrap', 'text-align': 'left', 'vertical-align': 'top'})
```
